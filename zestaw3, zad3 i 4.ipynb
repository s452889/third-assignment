```python
import os
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
try:
    from sklearn.metrics import plot_confusion_matrix
except(ImportError):
    from sklearn.utils.multiclass import unique_labels
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# zad 3
pd.set_option('display.max_columns', 500)
df = pd.read_csv("survey_results_public.csv")
df = df.dropna()
# replace values in 'Hobbyist' column on 1 and 0
df["Hobbyist"].replace({"Yes":1, "No":0}, inplace=True)
df["Student"].replace({"Yes, full-time":1, "Yes, part-time":1,
                       "No":0}, inplace=True)
df["Employment"].replace({"Employed full-time":1, "Employed part-time":0}, inplace=True)
df["BetterLife"].replace({"Yes":1, "No":0}, inplace=True)
df["YearsCodePro"].replace({"Less than 1 year":0}, inplace=True)
print(df["YearsCodePro"].unique())
df["OrgSize"].replace({"2-9 employees":1, "10 to 19 employees":2,
                       "20 to 99 employees":3, "100 to 499 employees":4,
                       "500 to 999 employees":5, "1,000 to 4,999 employees":6,
                       "5,000 to 9,999 employees":7, "10,000 or more employees":8}, inplace=True)
df.head(10)
```

```python
# zad 3
# percentage of occupancy and zero rule model accuracy
student_percentage = sum(df["BetterLife"]) / len(df)
print("Employment percentage is: " + str(round(student_percentage, 4)))
print("Zero rule model accuracy on set is: " + 
      str(round(1 - student_percentage, 4)))
```

```python
# zad 3
# Salary converted to annual USD salaries using the exchange 
# rate on 2019-02-01, assuming 12 working months and 50 working weeks.
clf = LogisticRegression()
X_train = df[['YearsCodePro']]
y_train = df.BetterLife
```

```python
# zad 3
# logistic regression classifier on one independent variable
clf.fit(X_train, y_train)
y_train_pred = clf.predict(X_train)
print(y_train_pred)
```

```python
# zad 3
# accurancy for training set
# logistic regression classifier on one independent variable
# Salary converted to annual USD salaries using the exchange 
# rate on 2019-02-01, assuming 12 working months and 50 working weeks.
accuracy_score(y_train, y_train_pred)
print("Training set accuracy for logisitic regression model "
      + "on 'ConvertedComp' variable:\n" + 
      str(accuracy_score(y_train, y_train_pred)))
conf_matrix = confusion_matrix(y_train, y_train_pred)
tn, fp, fn, tp = conf_matrix.ravel()
print(tn, fp, fn, tp)
specificity = tn / (tn+fp)
print("Training set specificity for logisitic regression model on 'ConvertedComp' variable: " + str(specificity))
sensitivity = tp / (tp+fn)
print("Training set sensitivity for logisitic regression model on 'ConvertedComp' variable: " + str(sensitivity))
# sensitivity = recall
```

```python
# zad 3
# logistic regression classifier on all but date independent variables
clf_all = LogisticRegression()
X_train_all = df[['YearsCodePro', 'Employment', 'Hobbyist', 'Student']]
clf_all.fit(X_train_all, y_train);
```

```python
# zad 3
# logistic regression classifier on all but date independent variables
y_train_pred_all = clf_all.predict(X_train_all)
```

```python
# zad 3
# logistic regression classifier on all but date independent variables
clf_all_accuracy = accuracy_score(y_train, y_train_pred_all)
print("Training set accuracy for logisitic regression model " +
      "on all but date variable: " + str(clf_all_accuracy))
```

```python
# zad 3
# logistic regression classifier on all but date independent variables
conf_matrix = confusion_matrix(y_train, y_train_pred_all)
tn_all, fp_all, fn_all, tp_all = conf_matrix.ravel()
print(tn_all, fp_all, fn_all, tp_all)
specificity_all = tn_all / (tn_all+fp_all)
print("Training set specificity for logisitic regression model " +
      "on all but date variable: " + str(specificity_all))
sensitivity_all = tp_all / (tp_all+fn_all)
print("Training set specificity for logisitic regression model " +
      "on all but date variable: " + str(sensitivity_all))
```
